```python
import numpy as np
from PIL import Image
from IPython.display import Image as show_image
from tensorflow.keras.applications.inception_resnet_v2 import InceptionResNetV2
from tensorflow.keras.preprocessing import image
from tensorflow.keras.applications.inception_resnet_v2 import preprocess_input, decode_predictions
```

```python
img  = Image.open('place-flying-sunset-sky.jpg').resize((299,299))
```

```python
img = np.array(img)
```

```python
img.shape
```

```python
print(img.ndim)
```

```python
img = img.reshape(-1,299,299,3)
```

```python
img.shape
```

```python
print(img.ndim)
```

```python
img = preprocess_input(img)
```

```python
incresv2_model = InceptionResNetV2(weights = 'imagenet', classes = 1000)
```

```python
print(incresv2_model.summary())
print(type(incresv2_model))
```

```python
show_image(filename='place-flying-sunset-sky.jpg')
```

```python
predictions = incresv2_model.predict(img)
print('Predicted Categories: ', decode_predictions(predictions, top=2)[0])
```
